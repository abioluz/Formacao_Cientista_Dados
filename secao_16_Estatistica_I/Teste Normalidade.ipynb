# Teste de Normalidade

```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

## Dados Normais

```python
# Criação de uma variável com dados em uma distribuição normal com a função rvs (100 elementos)
dados = norm.rvs(size=1000)
dados
```

```python
# histograma
plt.hist(dados, bins=20)
plt.show()
```

```python
# Geração de gráfico para verificar se a distribuição é normal
fig, ax = plt.subplots(1,2,figsize=(18, 6))

ax[0].hist(dados, bins=20)
stats.probplot(dados, fit=True, plot=ax[1])
plt.show()
```

```python
# execução do teste de Shapiro
# H0: Dados estão normalmente distribuidos
# p < 0.05 = rejeita a hipótese nula (não é normal)
# p > 0.05 = aceita a hipótese nula (é normal)

# Segundo argumento é o valor p, não há como rejeitar a hipotese nula

stats.shapiro(dados)
```

## Dados não Normais

```python
dados2 = skewnorm.rvs(4, size=1000)
dados2
```

```python
# Histograma
fig2, ax2 = plt.subplots(1,2,figsize=(18,6))
ax2[0].hist(dados2, bins=20)
stats.probplot(dados2, fit=True, plot=ax2[1])
plt.show()
```

```python
# execução do teste de Shapiro
# H0: Dados estão normalmente distribuidos
# p < 0.05 = rejeita a hipótese nula (não é normal)
# p > 0.05 = aceita a hipótese nula (é normal)

# Segundo argumento é o valor p, não há como rejeitar a hipotese nula

stats.shapiro(dados2)
```
