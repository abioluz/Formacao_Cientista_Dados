# Gráfico de Dispersão

Ideal para comparar 2 variáveis

```python
# importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregando base de dados
base = pd.read_csv('trees.csv')
display(base.shape)
display(base.head())
```

```python
# girth com volume
plt.scatter(base.Girth, base.Volume)
```

```python
# girth com height
plt.scatter(base.Girth, base.Height)
```

```python
# height com volume
plt.scatter(base.Height, base.Volume, marker='*')
```

```python
# histograma de volume

plt.hist(base.Volume)
```

```python
# imprimindo juntos
# Criação de figura, no qual os gráficos serão posicionados
plt.figure(1, figsize=(12,6))

# plt.subplot(numero de linhas,numero de colunas,posição)
plt.subplot(2,2,1)
plt.scatter(base.Girth, base.Volume)

plt.subplot(2,2,2)
plt.scatter(base.Girth, base.Height)

plt.subplot(2,2,3)
plt.scatter(base.Height, base.Volume, marker='*')

plt.subplot(2,2,4)
plt.hist(base.Volume)
```

```python
# imprimindo juntos
# Criação de figura, no qual os gráficos serão posicionados
plt.figure(1, figsize=(12,6))

# plt.subplot(numero de linhas, numero de colunas, posição)
plt.subplot(3,2,1)
plt.scatter(base.Girth, base.Volume)

plt.subplot(3,2,2)
plt.scatter(base.Girth, base.Height)

plt.subplot(3,2,3)
plt.scatter(base.Height, base.Volume, marker='*')

plt.subplot(3,2,4)
plt.hist(base.Volume)

plt.subplot(3,1,3)
plt.hist(base.Volume)
```
