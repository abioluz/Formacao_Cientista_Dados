# Gráfico de Dispersão

Ideal para comparar 2 variáveis

```python
# importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregando base de dados
base = pd.read_csv('trees.csv')
display(base.shape)
```

```python
base.head()
```

```python
# Gráficos de dispersão considerando o volume e a dispersão
plt.scatter(x=base.Girth, y=base.Volume, color='blue', facecolors='none', marker='*')
plt.title('Árvores')
plt.xlabel('Circunferência')
plt.ylabel('Volume')
```

```python
# Gráfico de linha considerádo o volume e o atributo "girth" - sem marker
plt.plot(base.Girth, base.Volume)
plt.title('Árvores')
plt.xlabel('Circunferência')
plt.ylabel('Volume')
```

```python
# Gráfico de dispersão co, 'afastamento' dos dados (jitter)
# O jitter cria uma variabilidade nos pontos proximos para ter uma visualizaçõa melhor. 
# Porem o gráfico fica com pontos randomizados

# fit_reg linha de tendência

sns.regplot(x=base.Girth, y=base.Volume, data=base, x_jitter=0.3, fit_reg=True)
```
